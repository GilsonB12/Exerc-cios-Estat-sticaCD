# Task

Você tem dados de emprego sobre o país X. Você foi solicitado a preparar uma tabela cruzada mostrando esses dados.
60% dos jovens de 18 a 25 anos estão empregados
85% das pessoas de 25 a 35 anos estão empregadas
5% das pessoas de 35 a 45 anos estão desempregadas
3% das pessoas de 45 a 55 anos estão desempregadas
3% das pessoas de 55 a 65 anos estão desempregadas
Todos os 65+ estão empregados. Nota: a definição de desempregado é: sem emprego, mas procurando ativamente por um. É provavelmente por isso que todos os 65+ estão empregados.


Crie uma tabela cruzada resumindo os dados que você recebeu.

Crie um gráfico de barras lado a lado (é chamado de gráfico de colunas agrupadas no Excel), para aprimorar visualmente seu resumo.


Por conta dessa frase "Todos os 65+ estão empregados. Nota: a definição de desempregado é: sem emprego, mas procurando ativamente por um. É provavelmente por isso que todos os 65+ estão empregados." estou considerando que quem tem +65 de anos está desempregado

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Definindo os dados
categorias = ['18-25', '25-35', '35-45', '45-55', '55-65', '65+']
empregados = np.array([60, 85, 5, 3, 3, 0])
desempregados = 100 - empregados
tabela_cruzada = pd.DataFrame({
    'Categoria': categorias,
    'Empregados (%)': empregados,
    'Desempregados (%)': desempregados
})

display(tabela_cruzada)

# Criando os 2 gráficos
fig, ax = plt.subplots(2, 1, figsize=(10, 12))
bar_width = 0.35

# Gráfico de barras 
index_corrigido = np.arange(len(categorias))
ax[0].bar(index_corrigido, empregados, bar_width, label='Empregados')
ax[0].bar(index_corrigido + bar_width, desempregados, bar_width, label='Desempregados')
ax[0].set_xlabel('Faixa Etária')
ax[0].set_ylabel('Porcentagem')
ax[0].set_title('Emprego por Faixa Etária')
ax[0].set_xticks(index_corrigido + bar_width / 2)
ax[0].set_xticklabels(categorias)
ax[0].legend()

# Gráfico de dispersão
for i, categoria in enumerate(categorias):
    ax[1].scatter([categoria] * 2, [empregados[i], desempregados[i]], label=categoria)

ax[1].set_xlabel('Faixa Etária')
ax[1].set_ylabel('Porcentagem')
ax[1].set_title('Dispersão de Emprego por Faixa Etária')
ax[1].legend(title="Categorias", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()

```
